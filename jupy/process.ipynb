```python
text = open('../../data/seg_txt.txt','r',encoding='utf-8').readlines()
word = set()
for t in text:
    for w in t.strip().split():
        word.add(w)
text.close()
```

```python
import numpy as np
```

```python
with open('../../data/wordvector.txt','r',encoding='utf-8') as fr:
    data = fr.readlines()
wordvector = data[1:]
dictionary2vec = {}
for line in wordvector:
    line = line.strip().split()
    vec = np.array(line[1:301]).astype(np.float32)
    vec = vec / (np.linalg.norm(vec) + 1e-6)
    dictionary2vec[line[0]] = vec
```

```python
word2id = {}
for (i, w) in enumerate(word):
    word2id[w] = i
```

```python
import numpy as np
id2v = {}
for w in word2id:
    if(w in dictionary2vec.keys()):
        id2v[word2id[w]] = dictionary2vec[w]
    else:
        vec = np.random.randn(300)
        vec = vec / (np.linalg.norm(vec) + 1e-6)
        id2v[word2id[w]] = vec
```
